```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Read the CSV
with open('node1sup.csv') as txt: 
    lines = [line for line in txt]
lines = lines[1:]

# Create an XYZH file
xyz = list()
for line in lines: 
    _, x, y, z, group, _, _, _ = line.split(",")
    xyz.append([x,y,z, group])
brain = np.array(xyz, dtype=np.float32)

# Switch dimensions around for better visualization
brain = brain[:,[0,2,1,3]]

# Rescale the brain so we can easily walk through it
brain[:,:3] *= 0.1

# Divide the last dimension for better colors. 
brain[:,3] /= 24.0

# Save to file ready to upload
brain.byteswap().tofile('brain.npy')
```
